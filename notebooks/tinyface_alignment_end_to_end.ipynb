```python
# 0. Đọc controller và tạo helper chạy từng stage
from pathlib import Path
import json, os
root=Path.cwd()
if not (root/'notebooks'/'tinyface_alignment_colab.ipynb').is_file():
    root=next((p for p in [Path('/content/Landmark-aligmenttest'),*root.parents] if (p/'notebooks'/'tinyface_alignment_colab.ipynb').is_file()),None)
if root is None: raise FileNotFoundError('Không tìm thấy repository Landmark-aligmenttest')
os.chdir(root)
controller=json.loads((root/'notebooks'/'tinyface_alignment_colab.ipynb').read_text())
stages=[c for c in controller['cells'] if c.get('cell_type')=='code']
def run_stage(index):
    source=''.join(stages[index]['source'])
    print(source.splitlines()[0] if source.strip() else f'stage {index}')
    exec(compile(source,f'tinyface_alignment_colab:stage-{index+1}','exec'),globals())
print('Repository:',root,'Stages:',len(stages))
```

```python
# 1. Bootstrap: paths và Python kernel
run_stage(0)
```

```python
# 2. Cài/kiểm tra gdown
run_stage(1)
```

```python
# 3. Tải và giải nén TinyFace
run_stage(2)
```
